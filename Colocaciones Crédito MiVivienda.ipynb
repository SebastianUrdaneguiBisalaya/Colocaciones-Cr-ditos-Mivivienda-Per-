## **Datasets**

```python
data_2018 = pd.read_csv(r'Data_NCMV_2018.csv')
data_2019 = pd.read_csv(r'Data_NCMV_2019.csv')
data_2020 = pd.read_csv(r'Data_NCMV_2020.csv')
data_2021 = pd.read_csv(r'Data_NCMV_2021.csv')
data_2022 = pd.read_csv(r'Data_NCMV_2022.csv')
```

## **Clustering con k-means utilizando con PCA**

```python
DATA_CHARACTER = DATA[["MONTO_CREDITO", "MONTO_CUOTA_INICIAL",
     "PLAZOS", "TASA", "MONTO_VALOR_VIVIENDA", "PRODUCTO", "DEPARTAMENTO", "TIPO_IFI"]]
DATA_CHARACTER["PLAZOS"] = DATA_CHARACTER["PLAZOS"].astype(float)
DATA_CHARACTER.head()
```

```python
DATA_CHARACTER.dtypes
```

```python
fig = plt.figure(figsize=(15,10))
numerical_columns = DATA_CHARACTER.select_dtypes(include=['float64']).columns
sns.heatmap(DATA_CHARACTER[numerical_columns].corr(), annot=True, 
            cmap="coolwarm")
plt.plot()
plt.show()
```

```python
object_columns = DATA_CHARACTER.select_dtypes(include=["object"]).columns
float_columns = DATA_CHARACTER.select_dtypes(include=["float64"]).columns

preprocessor = ColumnTransformer(
    transformers=[
        ('standard_scaler', StandardScaler(), float_columns),
        ('encoder', OneHotEncoder(), object_columns)
    ],
    remainder='passthrough'
)
DATA_CHARACTER_TRANSFORM = preprocessor.fit_transform(DATA_CHARACTER)
columns = preprocessor.get_feature_names_out(DATA_CHARACTER.columns)
DATA_CHARACTER_TRANSFORM = DATA_CHARACTER_TRANSFORM.todense()
DATA_CHARACTER_TRANSFORM_ = pd.DataFrame(DATA_CHARACTER_TRANSFORM,
                                        columns = columns)
DATA_CHARACTER_TRANSFORM_
```

```python
pca = PCA()
pca.fit_transform(DATA_CHARACTER_TRANSFORM_)
PCA_DATA_SCALED = pca.transform(DATA_CHARACTER_TRANSFORM_)
PCA_DATA_SCALED
```

```python
var = pca.explained_variance_ratio_
cum_var = np.cumsum(np.round(var, decimals=4)*100)
fig, ax = plt.subplots(figsize = (8,6), dpi = 80)
plt.title("Varianza Acumulada", size = 12)
plt.plot(cum_var, "r-x")
for i in ['bottom', 'left']:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(1.5)
right_side = ax.spines["right"]
right_side.set_visible(False)
top_side = ax.spines["top"]
top_side.set_visible(False)
ax.set_axisbelow(True)
ax.grid(color='gray', linewidth=1, axis='y', alpha=0.4)
plt.show()
```

```python
PCA_DATA_SCALED_STANDARD = pd.DataFrame(PCA_DATA_SCALED,
                                       columns = columns)
PCA_DATA_SCALED_STANDARD.drop(PCA_DATA_SCALED_STANDARD.columns[3:], 
                              axis=1,inplace=True)
PCA_DATA_SCALED_STANDARD
```

```python
fig, ax = plt.subplots(figsize = (8,6), dpi = 80)
plt.scatter(DATA["MONTO_CREDITO"], DATA["MONTO_CUOTA_INICIAL"],
           c = 'white', marker = 'o', edgecolor = "blue")
for i in ['bottom', 'left']:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(1.5)
right_side = ax.spines["right"]
right_side.set_visible(False)
top_side = ax.spines["top"]
top_side.set_visible(False)
ax.set_axisbelow(True)
ax.grid(color='gray', linewidth=1, axis='y', alpha=0.4)
plt.xlabel("Monto de crédito")
plt.ylabel("Monto de cuota inicial")
plt.title("Relación entre el monto de crédito y la cuota inicial", size = 12)
plt.show()
```

```python
sum_of_squared_distances = []
silhouette_scores = []
k = range(2, 8)
for i in k:
    print(i)
    km = KMeans(n_clusters = i)
    y_predict = km.fit_predict(PCA_DATA_SCALED_STANDARD)
    sum_of_squared_distances.append(km.inertia_)
    silhouette_scores.append(silhouette_score(PCA_DATA_SCALED_STANDARD, y_predict))
```

```python
fig, ax = plt.subplots()
visualizer = KElbowVisualizer(KMeans(), k=(1,8))
visualizer.fit(PCA_DATA_SCALED_STANDARD)
visualizer.show()
```

```python
fig, ax = plt.subplots(figsize = (8,6), dpi = 70)
ax = plt.gca()
plt.plot(k, sum_of_squared_distances, "bx-")
plt.xlabel("Grupos")
plt.ylabel("Inercia")
plt.title("Método del Codo", size = 14)
ax.set_facecolor("white")
for i in ['bottom', 'left']:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(1.5)
right_side = ax.spines["right"]
right_side.set_visible(False)
top_side = ax.spines["top"]
top_side.set_visible(False)
ax.set_axisbelow(True)
ax.grid(color='gray', linewidth=1, axis='y', alpha=0.4)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,6), dpi = 70)
plt.plot(k, silhouette_scores, "rx-")
plt.xlabel("K")
plt.ylabel("silhouette_scores")
plt.title("Silhouette Scores", size = 14)
ax.set_facecolor("white")
for i in ['bottom', 'left']:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(1.5)
right_side = ax.spines["right"]
right_side.set_visible(False)
top_side = ax.spines["top"]
top_side.set_visible(False)
ax.set_axisbelow(True)
ax.grid(color='gray', linewidth=1, axis='y', alpha=0.4)
plt.show()
```

```python
km = KMeans(n_clusters = 2, random_state = 0)
y_predict = km.fit_predict(PCA_DATA_SCALED_STANDARD)
print(silhouette_score(PCA_DATA_SCALED_STANDARD, y_predict))
DATA_CHARACTER["KMeans"] = y_predict
PCA_DATA_SCALED_STANDARD["KMeans"] = y_predict
```

```python
fig, ax = plt.subplots(figsize = (8,6), dpi = 80)
plt.scatter(PCA_DATA_SCALED_STANDARD[PCA_DATA_SCALED_STANDARD["KMeans"] == 0]["standard_scaler__MONTO_CREDITO"], 
            PCA_DATA_SCALED_STANDARD[PCA_DATA_SCALED_STANDARD["KMeans"] == 0]["standard_scaler__MONTO_CUOTA_INICIAL"],
           s = 100, c = "#9b2948", label = "Segmento 1")
plt.scatter(PCA_DATA_SCALED_STANDARD[PCA_DATA_SCALED_STANDARD["KMeans"] == 1]["standard_scaler__MONTO_CREDITO"], 
            PCA_DATA_SCALED_STANDARD[PCA_DATA_SCALED_STANDARD["KMeans"] == 1]["standard_scaler__MONTO_CUOTA_INICIAL"],
           s = 100, c = "#ffca7b", label = "Segmento 2")
plt.scatter(PCA_DATA_SCALED_STANDARD[PCA_DATA_SCALED_STANDARD["KMeans"] == 2]["standard_scaler__MONTO_CREDITO"], 
            PCA_DATA_SCALED_STANDARD[PCA_DATA_SCALED_STANDARD["KMeans"] == 2]["standard_scaler__MONTO_CUOTA_INICIAL"],
           s = 100, c = "#ff7251", label = "Segmento 3")
plt.scatter(km.cluster_centers_[:,0], 
           km.cluster_centers_[:,1],
           s = 400, c = "orange", label = "Centroides", marker = "*")
plt.legend()
plt.legend(scatterpoints = 1)
plt.grid()
plt.tight_layout()
for i in ['bottom', 'left']:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(1.5)
right_side = ax.spines["right"]
right_side.set_visible(False)
top_side = ax.spines["top"]
top_side.set_visible(False)
ax.set_axisbelow(True)
ax.grid(color='gray', linewidth=1, axis='y', alpha=0.4)
plt.title("Clustering con KMeans", size = 14)
plt.xlabel("Monto del crédito", size = 12)
plt.ylabel("Monto de la cuota inicial", size = 12)
plt.show()
```
